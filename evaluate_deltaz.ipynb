```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dz = [0.02, 0.04, 0.06, 0.08, 0.1, 0.12, 0.14, 0.16, 0.18, 0.2, 0.22, 0.24, 0.26, 0.28, 0.3, 0.32, 0.34, 0.36, 0.38, 0.4, 0.42, 0.44, 0.46, 0.48, 0.5, 0.52, 0.54, 0.56, 0.58, 0.6, 0.62, 0.64, 0.66, 0.68, 0.7, 0.72, 0.74, 0.76, 0.78]
```

```python
StoN = np.zeros(len(dz))
for i in range(0,len(dz)):
    StoN[i] = np.sqrt(np.loadtxt('./txtfiles/StoNstat_Blazek_LSST_DESI_deltaz='+str(dz[i])+'.txt'))
```

```python
plt.figure(figsize=[8,8])
plt.rc('font', family='serif', size=14)
plt.plot(dz, StoN, 'bo')
plt.xlabel('$\Delta z$')
plt.ylabel('Signal to Noise')
plt.title('Stastitical error only')
plt.savefig('./plots/deltaz_test_statonly.png')
plt.show()
```

```python
StoN_B = np.zeros(len(dz))
for i in range(0,len(dz)):
    StoN_B[i] = np.sqrt(np.loadtxt('./txtfiles/StoN_stat_sysB_Blazek_LSST_DESI_deltaz='+str(dz[i])+'.txt'))
```

```python
plt.figure(figsize=[8,8])
plt.rc('font', family='serif', size=14)
plt.plot(dz, StoN_B, 'mo')
plt.xlabel('$\Delta z$')
plt.ylabel('Signal to Noise')
plt.title('Stastitical error + systematics due to boost')
plt.savefig('./plots/deltaz_test_stat_sysB.png')
plt.show()
```
