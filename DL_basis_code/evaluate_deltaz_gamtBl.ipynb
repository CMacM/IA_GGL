```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dz = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]
```

```python
StoN = np.zeros(len(dz))
for i in range(0,len(dz)):
    StoN[i] = np.sqrt(np.loadtxt('./txtfiles/StoNstat_Blazek_LSST_DESI_deltaz='+str(dz[i])+'_gamtBl.txt'))
```

```python
plt.figure(figsize=[8,8])
plt.rc('font', family='serif', size=14)
plt.plot(dz, StoN, 'bo')
plt.xlim(0.0, 0.6)
plt.ylim(1,3)
plt.xlabel('$\Delta z$')
plt.ylabel('Signal to Noise')
plt.title('Stastitical error only')
plt.savefig('./plots/deltaz_test_statonly_gamtBl.png')
plt.show()
```

```python
StoN_B = np.zeros(len(dz))
for i in range(0,len(dz)):
    StoN_B[i] = np.sqrt(np.loadtxt('./txtfiles/StoN_stat_sysB_Blazek_LSST_DESI_deltaz='+str(dz[i])+'_gamtBl.txt'))
```

```python
plt.figure(figsize=[8,8])
plt.rc('font', family='serif', size=14)
plt.plot(dz, StoN_B, 'mo')
plt.xlim(0.0, 0.6)
plt.ylim(1,3)
plt.xlabel('$\Delta z$')
plt.ylabel('Signal to Noise')
plt.title('Stastitical error + systematics due to boost')
plt.savefig('./plots/deltaz_test_stat_sysB_gamtBl.png')
plt.show()
```
