```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
(z, COSMOS_dNdz, SDSS_dNDZ) = np.loadtxt('../txtfiles/DESI_redshifts.txt', unpack=True)
```

```python
plt.figure(figsize=(13,13))
plt.plot(z, COSMOS_dNdz, 'go')
plt.hold(True)
plt.plot(z, SDSS_dNDZ, 'mo')
```

```python
zeff_cosmos = sum(z[1:]*COSMOS_dNdz[1:]) / (sum(COSMOS_dNdz[1:]))
print "zeff, cosmos redshifts=", zeff_cosmos

zeff_SDSS = sum(z*SDSS_dNDZ) / (sum(SDSS_dNDZ))
print "zeff, sdss redshifts=", zeff_SDSS
```
