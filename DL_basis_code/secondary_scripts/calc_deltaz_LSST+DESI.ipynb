```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
import scipy.interpolate
```

```python
(z_ph, dNdzph) = np.loadtxt('../txtfiles/dNdz_ph_DESI_LSST.txt', unpack=True)
```

```python
Total = scipy.integrate.simps(dNdzph, z_ph)
print Total
interpdNdz = scipy.interpolate.interp1d(z_ph, dNdzph)
```

```python
delz = np.asarray([0.1,0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.56, 0.57, 0.58, 0.59, 0.6, 0.65])
```

```python
for i in range(0,len(delz)):
    zph_del = scipy.linspace(z_ph[0]+delz[i], z_ph[-1],1000)
    dNdz_del = interpdNdz(zph_del)
    total_del = scipy.integrate.simps(dNdz_del, zph_del)
    print "del=", delz[i], "frac=", total_del
```

```python
plt.figure(figsize=(13,13))
plt.plot(z_ph, dNdzph)
plt.hold(True)
plt.axvline(x=(0.77+0.57))
plt.show()
```
