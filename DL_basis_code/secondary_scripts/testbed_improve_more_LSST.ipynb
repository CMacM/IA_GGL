```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the wgg and wgp files both from running with survey = 'SDSS' with 'all' parameters changes to LSST 
# And for running with survey = 'LSST_DESI' with params.py with 'all' LSST params -> params_LSST_DESI.py 

(rp, wgp1h_SDSS_LSSTall) = np.loadtxt('../txtfiles/wgp_1h_survey=SDSS_rlim=25.3_LSSTall_fixbothdN.txt', unpack=True)
(rp, wgp1h_LSST_SDSSfile) = np.loadtxt('../txtfiles/wgp_1h_survey=LSST_DESI_rlim=25.3_SDSSfile_testah.txt', unpack=True)

plt.figure()
plt.loglog(rp, wgp1h_SDSS_LSSTall, 'mo', label='LSSTall')
plt.hold(True)
plt.loglog(rp, wgp1h_LSST_SDSSfile, 'go', label='SDSSfile')
plt.legend()
plt.show()
```
