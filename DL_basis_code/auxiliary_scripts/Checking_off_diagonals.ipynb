```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

This notebook checks whether off-diagonal elements of the Delta Sigma / gamma_t covariance matrices are sufficiently sub-dominant to diagonal elements to be ignored. (The off diagonal elements may be non-zero due to cosmic-variance terms.)

```python
# Import covariance matrix
SURVEY = 'SDSS'
observable = 'DelSig'
SAMPLE = 'A'

if observable=='DelSig':
    cov = np.loadtxt('../txtfiles/cov_DelSig_withCosmicVar_'+SURVEY+'_sample='+SAMPLE+'_rpts2500_lpts100000_fix_clgk_clkk.txt')
elif observable=='gammat':
    cov = np.loadtxt('../txtfiles/cov_gamt_withCosmicVar_'+SURVEY+'rpts2500_lpts100000_gg+const.txt')
```

```python
# Make colour plot

plt.figure(figsize=(13, 13))
plt.imshow(np.log10(np.abs(cov)), aspect=1, interpolation='None')
plt.colorbar()
plt.xlabel('rp', fontsize=20)
plt.ylabel('rp', fontsize=20)
plt.title('Log(|Covariance matrix|), DeltaSigma, SDSS, sample A', fontsize=20)
plt.show()
```
