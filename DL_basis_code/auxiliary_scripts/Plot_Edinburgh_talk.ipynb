```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
```

```python
zs = 0.303
a = 2.338
z = np.linspace(0, 1.2, 1000)
nofz_ = (z / zs)**(a-1) * np.exp( -0.5 * (z / zs)**2)
norm = scipy.integrate.simps(nofz_, z)
```

```python
plt.figure(figsize=(10,10))
plt.plot(z, nofz_ / norm, color='k',linewidth=2, label='Source distributions')
plt.hold()
plt.axvline(x=0.45, color='k', linestyle='--',linewidth=3, label='Photo-z split between samples')
plt.hold()
plt.axvline(x=0.28, color='m', linewidth=3, label='Mean lens z')
plt.xlabel('$z$', fontsize=40)
plt.ylabel('$\\frac{dN}{dz}$', fontsize=40)
plt.tick_params(axis='both', labelsize='30')
#plt.legend(fontsize=20)
#plt.show()
plt.tight_layout()
plt.savefig('/home/danielle/Dropbox/CMU/Conferences/UKTravelSummer2017/Edinburgh_talk/Blazek_demo_plot.png')
```
