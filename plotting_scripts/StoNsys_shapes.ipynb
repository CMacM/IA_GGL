```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
(fudge_Ncorr, Ratio) = np.loadtxt('../txtfiles/fudgeNcorr.txt', unpack=True)
```

```python
plt.figure(figsize=(13, 13))
plt.semilogy(fudge_Ncorr, Ratio, 'mo')
plt.xlabel('Fractional error', fontsize=20)
plt.ylabel('Ratio, S-to-N sys / stat', fontsize=20)
plt.savefig('../plots/StoNsysvstat_shapes.pdf')
plt.show()
```
