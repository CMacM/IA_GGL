```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(linewidth=120)
```

This notebook makes a 2D plot of the ration of statistical-only signal to noise from the Blazek et al case and from the shape-measurement case as a function of a and the percentage covaraince between methods.

```python
# Load StoN from Blazek method (only one value). We actually saved the square of what we want.
StoNsq_Blazek = np.loadtxt('../txtfiles/StoNstat_Blazek.txt')
StoN_Blazek = np.sqrt(StoNsq_Blazek)
print "StoN Blazek=", StoN_Blazek
```

```python
# Load vectors of a and the percentage covariance between methods
a = np.loadtxt('../txtfiles/a_StoNstat_extendeda.txt')
covperc = np.loadtxt('../txtfiles/cov_perc_StoNstat.txt')

a_inv = 1./a
print "(our a, a from Singh et al. 2016) =", zip(a, a_inv)
print "covariance percentage=", covperc

# Load the matrix of StoN values from the shape-measurement method at each pair of a / % covariance values
StoNsq_shapes = np.loadtxt('../txtfiles/StoNsq_shapes_a_extendeda.txt')
StoN_shapes = np.sqrt(StoNsq_shapes)
print "StoN_shapes="
print StoN_shapes
```

```python
# Make a 2D colour plot of the ratio of the S to N values.
# If the value plotted is greater than 1, the shape measurment does better for that combination of a and % cov.

StoNratio = StoN_shapes / StoN_Blazek
print "StoNratio="
print StoNratio

plt.figure(figsize=(13, 13))
plt.imshow(StoNratio, extent=[covperc[0], covperc[-1], a[-1], a[0]], aspect=1, interpolation='None')
plt.colorbar()
plt.xlabel('% covariance of methods', fontsize=20)
plt.ylabel('a (ratio of IA signal from methods)', fontsize=20)
plt.title('(Stat S/N, shapes method) / (Stat S/N, Blazek method)', fontsize=20)
#plt.show()
plt.savefig('../plots/StoN_2d_stat.pdf')
```
