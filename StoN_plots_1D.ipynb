```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
survey = 'SDSS'

(rp, StoN_Blazek) = np.loadtxt('./txtfiles/StoN_sysB_stat_Blazek_'+survey+'.txt', unpack=True)
(rp, StoN_shapes) = np.loadtxt('./txtfiles/StoN_sysB_stat_shapes_'+survey+'_a0pt3_rho0pt7.txt', unpack=True)
```

```python
fig_sub=plt.subplot(111)
plt.rc('font', family='serif', size=14)
fig_sub.scatter(rp ,StoN_Blazek, color='k', marker='o', s=100, label='Blazek et al. 2012')
fig_sub.hold(True)
fig_sub.scatter(rp ,StoN_shapes,color='m', marker='o', s=100, label='Multiple shape methods')
fig_sub.set_xscale("log")
fig_sub.set_xlabel('$r_p, \, {\\rm Mpc / h}$', fontsize=20)
fig_sub.set_ylabel('$\\frac{S}{N}$', fontsize=20)
fig_sub.set_ylim(0, 5.0)
fig_sub.set_xlim(0.05,20.)
fig_sub.tick_params(axis='both', which='major', labelsize=18)
fig_sub.tick_params(axis='both', which='minor', labelsize=18)
fig_sub.legend(loc='upper left')
#plt.title('SDSS sources + SDSS lenses. $a, \, \\rho=0.8$')
plt.savefig('./plots/StoN1d_SDSS.png')
plt.show()
```
