```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
```

```python
zL, dNdzL = np.loadtxt('./txtfiles/DESY1_quantities_fromSara/z_dNdz_lenses.dat', unpack=True)
```

```python
dNdz_G = np.exp(-(zL-0.37)**2/(2*0.005**2))
```

```python
norm_G = scipy.integrate.simps(dNdz_G, zL)
norm = scipy.integrate.simps(dNdzL, zL)
```

```python
plt.figure()
plt.plot(zL,dNdz_G/norm_G)
plt.plot(zL, dNdzL/norm)
plt.show()
```

```python
save_G = np.column_stack((zL, dNdz_G))
np.savetxt('./txtfiles/DESY1_quantities_fromSara/z_dNdz_lenses_narrow_test.dat', save_G)
```
