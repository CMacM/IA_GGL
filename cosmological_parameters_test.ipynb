```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pars = [[0.3,72.0], [0.3, 70.0], [0.3, 68.0], [0.28, 70.0], [0.28, 68.0]]
```

```python
gammas = [0]*len(pars); thetas = [0]*len(pars)
for i in range(0,len(pars)):
    thetas[i], gammas[i] = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_DESY1_assumed_OmM='+str(pars[i][0])+'_H0='+str(pars[i][1])+'.txt', unpack=True)
```

```python
plt.figure()
for i in range(0,len(pars)):
    plt.loglog(thetas[i], gammas[i], 'o', label='OmM='+str(pars[i][0])+'_H0='+str(pars[i][1]))
plt.ylabel('$\gamma_{IA}(\\theta)$')
plt.xlabel('$\\theta$, arcmin')
plt.title('No IA, vary cosmological parameters')
plt.ylim(10**(-6),10**(-3))
plt.legend()
plt.savefig('./vary_cosmological_parameters.pdf')
```
