```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Look at the S/N or frac err on the lensing quantity itself
SURVEY = 'SDSS'
(rp, gamt_fracerr_asc) = np.loadtxt('../txtfiles/gt_fracerr_survey='+SURVEY+'_sample=assoc.txt', unpack=True)
(rp, delsig_fracerr_asc) = np.loadtxt('../txtfiles/DS_fracerr_survey='+SURVEY+'_sample=A.txt', unpack=True)
(rp, gamt_fracerr_B) = np.loadtxt('../txtfiles/gt_fracerr_survey='+SURVEY+'_sample=B.txt', unpack=True)
(rp, delsig_fracerr_B) = np.loadtxt('../txtfiles/DS_fracerr_survey='+SURVEY+'_sample=B.txt', unpack=True)
```

```python
plt.figure(figsize=[8,8])
plt.loglog(rp, 1./gamt_fracerr_asc, 'm', label='gammat assoc')
plt.hold(True)
plt.loglog(rp, 1./delsig_fracerr_asc, 'g', label='Delta Sigma A')
plt.hold(True)
plt.loglog(rp, 1./gamt_fracerr_B, 'm', linestyle = '--', label='gammat B')
plt.hold(True)
plt.loglog(rp, 1./delsig_fracerr_B, 'g', linestyle = '--', label='Delta Sigma B')
plt.ylabel('S/N', fontsize=30)
plt.xlabel('$r_p$', fontsize=30)
plt.legend(fontsize=14)
#plt.savefig('../plots/S_to_N_gamma_DelSig.png')
plt.show()
```

```python
SURVEY='SDSS'
# Look now as the S/N on gammaIA from each case
(rp, ston_gIA_gamt) = np.loadtxt('../txtfiles/StoN_stat_sysB_Blazek_gamt_'+SURVEY+'_deltaz=0.17_rlim=22.0.txt', unpack=True)
(rp, ston_gIA_DS) = np.loadtxt('../txtfiles/StoN_sysb_stat_Blazek_'+SURVEY+'_deltaz=0.17_rlim=22.0.txt', unpack=True)

plt.figure(figsize=[8,8])
plt.loglog(rp, ston_gIA_gamt, 'mo', label='gammat')
plt.hold(True)
plt.loglog(rp, ston_gIA_DS, 'go', label='Delta Sigma')
plt.legend()
plt.show()
```
