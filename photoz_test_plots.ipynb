```python
import numpy as np
import matplotlib.pyplot as plt
import pyccl as ccl
import params_LSST_DESI as pa
```

```python
# Translate rp to arcmin so I can easily compare to Sara

def rp_to_arcmin(rp):
    cosmo = ccl.Cosmology(Omega_c = pa.OmC, Omega_b = pa.OmB, 
                          h = (pa.HH0/100.), sigma8 = pa.sigma8, n_s=pa.n_s) 
    chieff = ccl.comoving_radial_distance(cosmo,1./(1.+pa.zeff))
    rads = rp / chieff
    arcmins = rads * 180 / np.pi *60
    
    return arcmins
```

```python
# Load gamma IA:

rp_IA_fid, gamIA_fid = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_0percent.txt',unpack=True)
#rp_IA_3, gamIA_3 = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_3percent.txt',unpack=True)
rp_IA_5, gamIA_5 = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_5percent.txt',unpack=True)
rp_IA_10, gamIA_10 = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_10percent.txt',unpack=True)
rp_IA_m5, gamIA_m5 = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_-5percent.txt',unpack=True)
rp_IA_m10, gamIA_m10 = np.loadtxt('./txtfiles/photo_z_test/gamma_IA_est_SDSS_dndz_-10percent.txt',unpack=True)
```

```python
theta_IA_fid = rp_to_arcmin(rp_IA_fid)
#theta_IA_3 = rp_to_arcmin(rp_IA_3)
theta_IA_5 = rp_to_arcmin(rp_IA_5)
theta_IA_10 = rp_to_arcmin(rp_IA_10)
theta_IA_m5 = rp_to_arcmin(rp_IA_m5)
theta_IA_m10 = rp_to_arcmin(rp_IA_m10)
```

```python
plt.figure()
plt.loglog(theta_IA_fid, np.abs(gamIA_fid), '+', markersize=6, label='fiducial')
#plt.loglog(theta_IA_3, np.abs(gamIA_3), '--', label='dNdz 3% off')
plt.loglog(theta_IA_5, np.abs(gamIA_5), '+', label='true dNdz pars +5%')
plt.loglog(theta_IA_10, np.abs(gamIA_10), '+', label='true dNdz pars +10%')
plt.loglog(theta_IA_m5, np.abs(gamIA_m5), '+', label='true dNdz pars -5%')
plt.loglog(theta_IA_m10, np.abs(gamIA_m10), '+', label='true dNdz pars -10%')
plt.ylim(10**(-5), 0.001)
plt.ylabel('$\gamma_{IA}$', fontsize=16)
plt.xlabel('$\\theta$, arcmin', fontsize=16)
plt.title('SDSS', fontsize='16')
plt.legend()
plt.savefig('./plots/photoz_test/SDSS_dNdz_test.pdf')
plt.show()
```
