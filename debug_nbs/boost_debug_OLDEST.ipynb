```python
z_l, dndz_l, zleff = zed.get_dndz_spec('lens')
print(np.max(z_l))
a_l = np.flip(1. / (1. + z_l))

k_arr = np.geomspace(1E-4, 5E2, 3000)

Pk_1h2h = halo.get_Pk2D(corr='gg', k_arr=k_arr, a_arr=a_l, onehalo=True)
Pk_2h = halo.get_Pk2D(corr='gg', k_arr=k_arr, a_arr=a_l, onehalo=False)
```

```python
zl = 299

r = np.logspace(np.log10(0.15), np.log10(10), 1000) * (sp.pa.HH0 / 100.)

xi_gg_1h2h = ccl.correlations.correlation_3d(zed.cosmo_SRD, np.flip(a_l[zl]), 
                                           r / (sp.pa.HH0 / 100.), Pk_1h2h)

xi_gg_2h = ccl.correlations.correlation_3d(zed.cosmo_SRD, np.flip(a_l[zl]), 
                                           r / (sp.pa.HH0 / 100.), Pk_2h)
```

```python
plt.figure(figsize=[6,5])
plt.plot(k_arr, Pk_1h2h.eval(k_arr, a_l[zl], cosmo=zed.cosmo_SRD), 
         color='magenta')
plt.plot(k_arr, Pk_2h.eval(k_arr, a_l[zl], cosmo=zed.cosmo_SRD), color='orange')
plt.title(r'$z_{l} = %g$'%(np.flip(z_l[zl])), fontsize=15)
plt.xlabel(r'$k / Mpc^{-1}$', fontsize=15)
plt.ylabel(r'$P_{gg}(k)$', fontsize=15)
plt.xscale('log')
plt.yscale('log')
plt.legend(['1h + 2h', '2h only'])
plt.ylim([1,10**6])
```
